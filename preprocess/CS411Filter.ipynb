```python
import pandas as pd
```

```python
df = pd.read_csv("US_youtube_trending_data.csv", encoding='utf-8')
df.head()
```

```python
df.shape
```

```python
df = df.drop(columns=['channelId', 'comments_disabled', 'ratings_disabled', 'comment_count'], axis=1)
```

```python
df['publishedAt'] = df['publishedAt'].str[:10].astype(str)
df['trending_date'] = df['trending_date'].str[:10].astype(str)
df['description'] = df['description'].str[:100]
df['tags'] = df['tags'].str[:250]
```

```python
print(df['video_id'].str.len().max())
print(df['title'].str.len().max())
print(df['publishedAt'].str.len().max())
print(df['channelTitle'].str.len().max())
print(df['trending_date'].str.len().max())
print(df['tags'].str.len().max())
print(df['thumbnail_link'].str.len().max())
print(df['description'].str.len().max())
```

```python
df.to_csv("youtube_trending_data_filter.csv", encoding="utf-8", index=False)
```
